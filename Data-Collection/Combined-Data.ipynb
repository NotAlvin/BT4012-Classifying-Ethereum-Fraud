```python
import pandas as pd
df_mined = pd.read_csv('../Data/address_data_e.csv')
print(df_mined.shape)
print(df_mined.head())
```

```python
df_kaggle = pd.read_csv('../Data/address_data_k.csv')
df_kaggle.drop(['Index','Unnamed: 0'], axis=1, inplace=True)
```

```python
erc_columns = [' Total ERC20 tnxs', ' ERC20 total Ether received',
       ' ERC20 total ether sent', ' ERC20 total Ether sent contract',
       ' ERC20 uniq sent addr', ' ERC20 uniq rec addr',
       ' ERC20 uniq sent addr.1', ' ERC20 uniq rec contract addr',
       ' ERC20 avg time between sent tnx', ' ERC20 avg time between rec tnx',
       ' ERC20 avg time between rec 2 tnx',
       ' ERC20 avg time between contract tnx', ' ERC20 min val rec',
       ' ERC20 max val rec', ' ERC20 avg val rec', ' ERC20 min val sent',
       ' ERC20 max val sent', ' ERC20 avg val sent',
       ' ERC20 min val sent contract', ' ERC20 max val sent contract',
       ' ERC20 avg val sent contract', ' ERC20 uniq sent token name',
       ' ERC20 uniq rec token name', ' ERC20 most sent token type',
       ' ERC20_most_rec_token_type']
df_kaggle.drop(erc_columns,axis=1,inplace=True)
df_kaggle.head()
```

```python
df_mined.loc[:,['Avg min between sent tnx', 'Avg min between received tnx',
       'Time Diff between first and last (Mins)']] = df_mined.loc[:,['Avg min between sent tnx', 'Avg min between received tnx',
       'Time Diff between first and last (Mins)']].round(2)
```

```python
print(df_mined.head())
print(df_mined.shape)
```

```python
print(df_kaggle.head())
print(df_kaggle.shape)
```

```python
dfmerged = pd.merge(df_mined,df_kaggle,how='outer')
print(dfmerged.shape)
dfmerged.head()
```

```python
cols_to_drop = ['total ether sent contracts', 'max val sent to contract', 'Received Tnx', 'Sent tnx', 'total Ether sent', 'min value sent to contract','avg value sent to contract', 'Number of Created Contracts', 'max val sent', 'Unique Sent To Addresses']
```

```python
# dfmerged.drop(erc_columns, axis=1, inplace=True)
dfmerged.drop(cols_to_drop, axis=1, inplace=True)
print(dfmerged.shape)
dfmerged.head()
```

```python
dfmerged.to_csv('../Data/address_data_combined.csv', index=False)
```
